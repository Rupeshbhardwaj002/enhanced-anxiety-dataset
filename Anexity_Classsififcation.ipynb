```python
import pandas as pd
url ='https://raw.githubusercontent.com/Rupeshbhardwaj002/enhanced-anxiety-dataset/refs/heads/main/enhanced_anxiety_dataset.csv'
df = pd.read_csv(url)
df.head()

```

```python
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df = df.dropna(subset = ['Age'])
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
df['Occupation'] = LabelEncoder().fit_transform(df['Occupation'])
# df['Recent Major Life Event'] = OneHotEncoder().fit_transform(df['Recent Major Life Event'])
```
